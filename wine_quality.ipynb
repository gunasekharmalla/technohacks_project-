```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"D:\archive\winequality-red.csv")
print(dataset)
```

```python
column_n = ['volatile acidity'  ,'citric acid'  ,'residual sugar'  ,'chlorides' ,'density'    ,'sulphates' ,'alcohol'  ,'free sulfur dioxide'  ,'total sulfur dioxide'  ,'fixed acidity'  ,'pH'  ,'quality']
dataset = dataset.reindex(columns=column_n)
X = dataset.iloc[:,0:-1].values
y = dataset.iloc[:,-1].values
dataset
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(X,y,test_size = 0.3, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train_x[:,6:] = sc.fit_transform(train_x[:,6:])
test_x[:,6:] = sc.transform(test_x[:,6:])
```

```python
from sklearn.ensemble import RandomForestRegressor
Reg = RandomForestRegressor(n_estimators = 100, random_state = 1)
Reg.fit(train_x,train_y)
```

```python
y_pred = Reg.predict(test_x)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(test_y,y_pred))
```
